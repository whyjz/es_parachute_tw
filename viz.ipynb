```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import hvplot.pandas
```

```python
# font = {'family' : 'normal',
#         'weight' : 'bold',
#         'size'   : 22}
font = {'size'   : 20}
matplotlib.rc('font', **font)
matplotlib.rc('legend', fontsize=20)
axes_settings = {'linewidth'   : 2}
matplotlib.rc('axes', **axes_settings)
```

```python
fname = '/home/jovyan/Projects/Workbench/TW_parachute/計算.xlsx'
df = pd.read_excel(fname, sheet_name=['B1', 'B2', 'C1', 'C2'])
df_b1, df_b2, df_c1, df_c2 = [df[key] for key in ['B1', 'B2', 'C1', 'C2']]
```

```python
au1_co = pd.concat([df_b1['AU1_CO'], df_b2['AU1_CO'], df_c1['AU1_CO'], df_c2['AU1_CO']], ignore_index=True)
cites = pd.concat([df_b1['TC'], df_b2['TC'], df_c1['TC'], df_c2['TC']], ignore_index=True)
```

```python
cites
```

```python
tw_mask = au1_co == 'TAIWAN'
tw_cites = cites[tw_mask]
nontw_cites = cites[~tw_mask]
```

```python
tw_cites = tw_cites.dropna()
tw_cites = tw_cites.sort_values(ascending=False)
nontw_cites = nontw_cites.dropna()
nontw_cites = nontw_cites.sort_values(ascending=True)
```

```python
# len(tw_cites)
```

```python
# range(10)
```

```python
fig, ax0 = plt.subplots(1, 1, figsize=(20, 8))
plt.bar(range(-len(nontw_cites), 0), nontw_cites, width=1)
plt.bar(range(len(tw_cites)), tw_cites, width=1)
```

```python
fig, ax0 = plt.subplots(1, 1, figsize=(20, 8))
plt.bar(range(len(nontw_cites)-1, -1, -1), nontw_cites, width=1, alpha=0.7, label='First author affiliation is NOT TAIWAN')
plt.bar(range(len(tw_cites)), tw_cites, width=1, alpha=0.7, label='First author affiliation is TAIWAN')
plt.xlim(-1, 300)
plt.xlabel('Article order, sorted by total citations')
plt.ylabel('Total citations')
plt.legend();
```
